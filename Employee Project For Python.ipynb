```python
print("link for dataset:  https://www.kaggle.com/datasets/tawfikelmetwally/employee-dataset")


# Install required libraries (uncomment if not installed)
# !pip install pandas numpy matplotlib seaborn

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
from IPython.display import display



# Step 1: Import Dataset
print("Please upload your dataset (CSV format).")
uploaded = files.upload()

# Load the uploaded file into a DataFrame
file_name = list(uploaded.keys())[0]  # Get the uploaded file name
df = pd.read_csv(file_name)
print("\nDataset loaded successfully!")
print(f"Dataset Shape: {df.shape}")

# Display initial rows
display(df.head())



# Step 2: Inspect and Clean Data
# Inspect for missing values
print("\nInspecting for missing values...")
print(df.isnull().sum())

# Handle missing values
print("\nHandling missing values...")
for column in df.columns:
    if df[column].isnull().sum() > 0:
        if df[column].dtype == 'object':  # Categorical columns
            df[column].fillna(df[column].mode()[0], inplace=True)
        else:  # Numerical columns
            df[column].fillna(df[column].mean(), inplace=True)
print("Missing values handled using mean (numerical) or mode (categorical).")

# Verify missing values are handled
print("\nMissing values after handling:")
print(df.isnull().sum())

# Check and handle duplicates
print("\nChecking for duplicates...")
duplicates = df.duplicated().sum()
if duplicates > 0:
    print(f"Found {duplicates} duplicate rows. Removing duplicates...")
    df = df.drop_duplicates()
else:
    print("No duplicates found.")



# Step 3: Handle Outliers (numerical columns)
print("\nHandling outliers...")
numerical_columns = df.select_dtypes(include=[np.number]).columns
for column in numerical_columns:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()
    if outliers > 0:
        print(f"{outliers} outliers detected in column '{column}'. Capping them...")
        df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
        df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

# Ensure correct data types
print("\nEnsuring correct data types...")
print(df.dtypes)



# Step 4: Exploratory Data Analysis (EDA)
print("\nPerforming EDA...")

# Descriptive statistics
print("\nDescriptive Statistics:")
display(df.describe())

# Histograms for numerical data
print("\nVisualizing numerical distributions...")
df.hist(figsize=(15, 10))
plt.tight_layout()
plt.show()

# Box plots for outliers
print("\nVisualizing outliers using box plots...")
for column in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(df[column])
    plt.title(f"Box Plot - {column}")
    plt.show()

# Scatter plots for numerical relationships
print("\nVisualizing relationships using scatter plots...")
sns.pairplot(df.select_dtypes(include=[np.number]), diag_kind="kde")
plt.show()

# Correlation heatmap
print("\nCorrelation heatmap:")
numerical_columns = df.select_dtypes(include=[np.number])  # Select only numeric columns
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_columns.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Bar charts for categorical variables
categorical_columns = df.select_dtypes(include=['object']).columns
if len(categorical_columns) > 0:
    print("\nVisualizing categorical distributions...")
    for column in categorical_columns:
        plt.figure(figsize=(8, 4))
        sns.countplot(
            data=df,
            x=column,
            order=df[column].value_counts().index,
            palette="Set3"  # Use a Seaborn palette to add color differentiation
        )
        plt.title(f"Bar Chart - {column}")
        plt.xticks(rotation=45)
        plt.show()


# Step 5: Insights and Conclusion
print("\nInsights and Conclusion:")

# Example Insight: Correlation Analysis
if len(numerical_columns.columns) > 1:
    correlation_matrix = numerical_columns.corr()
    strong_correlations = correlation_matrix[(correlation_matrix > 0.7) & (correlation_matrix < 1)].stack()
    if not strong_correlations.empty:
        print("\nStrong Positive Correlations (above 0.7):")
        for index, value in strong_correlations.iteritems():
            print(f"- {index[0]} and {index[1]}: {value:.2f}")
    else:
        print("\nNo strong positive correlations (above 0.7) found.")

    negative_correlations = correlation_matrix[correlation_matrix < -0.7].stack()
    if not negative_correlations.empty:
        print("\nStrong Negative Correlations (below -0.7):")
        for index, value in negative_correlations.iteritems():
            print(f"- {index[0]} and {index[1]}: {value:.2f}")
    else:
        print("\nNo strong negative correlations (below -0.7) found.")

# Example Insight: Categorical Data Distribution3w
if len(categorical_columns) > 0:
    print("\nCategorical Variables Summary:")
    for column in categorical_columns:
        print(f"- {column}:")
        print(df[column].value_counts())

# Example Insight: Numerical Data Distribution
print("\nNumerical Variables Summary:")
for column in numerical_columns.columns:
    mean_value = df[column].mean()
    median_value = df[column].median()
    std_dev = df[column].std()
    print(f"- {column}: Mean = {mean_value:.2f}, Median = {median_value:.2f}, Std Dev = {std_dev:.2f}")

# Hypothesis or Insight Example
if 'price' in df.columns and 'sales' in df.columns:
    correlation = df['price'].corr(df['sales'])
    if correlation < 0:
        print(f"\nHypothesis: There is a negative correlation between 'price' and 'sales' ({correlation:.2f}). Higher prices might lead to lower sales.")
    else:
        print(f"\nHypothesis: There is a positive correlation between 'price' and 'sales' ({correlation:.2f}).")

# Next Steps
print("\nNext Steps:")
print("- Explore feature engineering to improve dataset quality (e.g., creating new features from existing data).")
print("- Apply advanced modeling techniques if predictive analysis is required.")
print("- Investigate potential external datasets to enrich the analysis.")

```
